```python
from tmdbv3api import TMDb,Movie
import json
tmdb_movie=Movie()
import requests
tmdb = TMDb()
tmdb.api_key ='480f60d18ac924348ee9d81ed6b76628'
```

```python
def genres(x):
    lis=[]
    j=0
    result=tmdb_movie.search(x)
    movie_id=result[0].id
    response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id,tmdb.api_key))
    data_json=response.json()
    if data_json['genres']:
        j+=1
        print(j)
        for i in range(len(data_json['genres'])):
            lis.append(data_json['genres'][i]['name'])
        out=" ".join(lis)
        return out
    else:
        return " "
```

```python
df['genres']=df['Movie_title'].apply(genres)
```

```python
def director(x):
    lis=x.split(';')[0]
    print(lis)
    if len(lis)>1:
        director=lis.split('(')
        director=director[0]
    return director
    
    
```

```python
df['director_name']=df['Cast and crew'].apply(director)
```

```python
def actor_1(x):
    lis=x.split(';')
    if len(lis)>2:
        actors=lis[2]
        a=actors.split(',')
        return a[0]
    else:
        return ' '
```

```python
def actor_2(x):
    lis=x.split(';')
    if len(lis)>2:
        actors=lis[2]
        a=actors.split(',')
        if len(a)>1:
            return a[1]
        else:
            return " "
    else:
        return ' '
```

```python
def actor_3(x):
    lis=x.split(';')
    if len(lis)>2:
        actors=lis[2]
        a=actors.split(',')
        if len(a)>2:
            return a[2]
        else:
            return ' '
    else:
        return ' '
```

```python
df['actor_1_name']=df['Cast and crew'].apply(actor_1)
df['actor_2_name']=df['Cast and crew'].apply(actor_2)
df['actor_3_name']=df['Cast and crew'].apply(actor_3)
```
