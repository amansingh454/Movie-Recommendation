```python
from ast import literal_eval
```

```python
merged_data['cast']=merged_data['cast'].apply(literal_eval)
```

```python
def actor_1(x):
    if len(x)==0:
         return " "
    else:
        if 'name' in x[0].keys():
            return x[0]['name']
        else:
            return " "
```

```python
def actor_2(x):
    if len(x)==0:
         return " "
    else:
        if len(x)>1:
             if 'name' in x[1].keys():
                return x[1]['name']
             else:
                return x[3]['name']
    return " "
```

```python
def actor_3(x):
    if len(x)==0:
        return " "
    else :
        if len(x)>2:
            if 'name' in x[2].keys():
                return x[2]['name']
            else:
                return x[4]['name']
    return " "
```

```python
merged_data['actor_1']=merged_data['cast'].apply(actor_1)
merged_data['actor_2']=merged_data['cast'].apply(actor_2)
merged_data['actor_3']=merged_data['cast'].apply(actor_3)
```

```python
def director(x):
    for i in x:
        if i['job']=='Director':
            return i['name']
        else:
            return ' ' 
    return ' '
```

```python
merged_data['crew']=merged_data['crew'].apply(literal_eval)
```

```python
merged_data['Director']=merged_data['crew'].apply(director)
```

```python
def genres(x):
    s=' '
    i=0
    if len(x)==0:
        return s
    else:
        while i<len(x):
            if 'name' in x[i].keys():
                s+=" "+x[i]['name']
                i+=1
    return s
```

```python
merged_data['genre']=merged_data['genres'].apply(genres)
```
